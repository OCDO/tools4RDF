```python
s = System.create.element.Si(graph = kg)
```

```python
s.write.file('qe_run/pw.si.scf.in', 
             format='quantum-espresso',
             copy_from='qe_run/pw.si.scf.ref',
             pseudo_files=['qe_run/Si.pbe-n-rrkjus_psl.1.0.0.UPF'])
```
